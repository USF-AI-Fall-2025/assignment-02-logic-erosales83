```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('gallstone.csv')
df.describe()
df.info()
```

```python
gallstone_counts = df['Gallstone Status'].value_counts()
plt.bar(['0', '1'], gallstone_counts)
plt.xticks([0,1], ['No Gallstone', 'Yes Gallstone'])
plt.ylabel('Count')
plt.title('Gallstone Status')
plt.show()
```

```python
counts = df['Gender'].value_counts()
plt.bar(['0', '1'], counts)
plt.xticks([0,1], ['Male', 'Female'])
plt.ylabel('Count')
plt.title('Gender')
plt.show()
```

```python
plt.hist(df['Glucose'])
plt.xlabel('Glucose Level')
plt.ylabel('Range')
plt.title('Glucose Levels')
plt.show()
```
